```python
# Q1: What is NumPy, and why is it widely used in Python?
'''
NumPy (Numerical Python) is a fundamental library for numerical computing in Python.
It provides:
- The ndarray: a fast, memory-efficient, N-dimensional array object.
- Vectorized operations for element-wise arithmetic and linear algebra.
- A large collection of mathematical functions (linear algebra, FFT, random, etc.).
Why widely used:
- Performance: implemented in C, operations on arrays are much faster than Python loops.
- Convenience: concise syntax for numerical tasks, broadcasting, slicing.
- Ecosystem: many scientific libraries (Pandas, SciPy, scikit-learn, TensorFlow) rely on NumPy arrays.
- Memory efficiency: contiguous memory and efficient storage for numeric types.

'''
```

```python
# Q2: How does broadcasting work in NumPy?
'''
Broadcasting is a set of rules NumPy uses to perform operations on arrays of different shapes.
Rules summary:
1. If arrays differ in rank, prepend 1s to the smaller shape.
2. Arrays are compatible if each dimension is equal or one of them is 1.
3. Dimensions with size 1 are 'stretched' (without copying data) to match the other array.
Example:
- (3,4) array + (4,) vector -> vector is treated as (1,4) then broadcast to (3,4).
Benefits:
- Avoids explicit replication, keeps memory use low and code concise.

'''
```

```python
# Q3: What is a Pandas DataFrame?
'''
A Pandas DataFrame is a 2-dimensional labeled data structure with columns of potentially different types.
Features:
- Row and column labels (index and columns).
- Operations: slicing, joining, groupby, aggregation, pivoting, time-series support.
- Designed for tabular data (like spreadsheets or SQL tables).

'''
```

```python
# Q4: Explain the use of the groupby() method in Pandas.
'''
groupby() splits the DataFrame into groups based on keys (one or more columns), then allows applying aggregation,
transformation, or filtering to each group.
Common steps: split -> apply -> combine.
Use cases:
- Compute summary statistics per group (mean, sum, count).
- Apply custom functions to groups.
- Transform data within group without reducing dimensionality.

'''
```

```python
# Q8: What does the term 'vectorized operation' mean in NumPy?
'''
Vectorized operations apply an operation to entire arrays (or large chunks) at once rather than element-by-element in Python loops.
Advantages:
- Uses optimized C loops under the hood.
- Much faster and more readable code.
Example: arr * 2 multiplies every element of arr without an explicit Python loop.

'''
```

```python
# Q21: Why is NumPy’s array slicing faster than Python’s list slicing?
'''
- NumPy slicing returns a view (no data copy) and operations are implemented in optimized C loops.
- Lists are collections of Python objects; slicing copies references and element-wise operations require Python-level iteration.
Thus numerical operations on NumPy slices avoid Python overhead.

'''
```

```python
# Practical 4: Calculate correlation matrix using Seaborn and visualize it with a heatmap.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df = pd.DataFrame({
    'A':[1,2,3,4,5],
    'B':[2,1,4,3,5],
    'C':[5,4,3,2,1]
})
corr = df.corr()
print("Correlation matrix:\n", corr)
plt.figure(figsize=(5,4))
sns.heatmap(corr, annot=True)
plt.title("Correlation heatmap")
plt.show()
```
